```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['gender'] = le.fit_transform(df['gender'])
```

```python
df.info()
```

```python
df['city_tier'] = le.fit_transform(df['city_tier'])
```

```python
df['shopping_preference'] = le.fit_transform(df['shopping_preference'])
```

```python
df.info()
```

```python
y = df['shopping_preference']
x = df.drop('shopping_preference', axis=1)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
modelLR = LogisticRegression()
```

```python
modelLR.fit(x_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, f1_score
y_predLR = modelLR.predict(x_test)
acc_LR = accuracy_score(y_test, y_predLR)
print(acc_LR)
print(confusion_matrix(y_test, y_predLR))
print(f1_score(y_test, y_predLR, average='weighted'))
```

```python
modelRF = RandomForestClassifier()
modelRF.fit(x_train, y_train)
y_predRF = modelRF.predict(x_test)
acc_RF = accuracy_score(y_test, y_predRF)
print(acc_RF)
print(f1_score(y_test, y_predRF, average='weighted'))
```

```python
modelXG = XGBClassifier()
modelXG.fit(x_train, y_train)
y_predXG = modelXG.predict(x_test)
acc_XG = accuracy_score(y_test, y_predXG)
print(acc_XG)
print(f1_score(y_test, y_predXG, average='weighted'))
```

```python
accuracy = {
    "Logistic Regression": acc_LR,
    "Random Forest": acc_RF,
    "XGBoost": acc_XG
}
df_acc = pd.DataFrame(list(accuracy.items()), columns=['Model', 'Accuracy'])
```

```python
import seaborn as sns
sns.barplot(x='Model', y='Accuracy', data=df_acc)
```

```python
from imblearn.over_sampling import SMOTE
smote = SMOTE(random_state=0)
X_train, Y_train = smote.fit_resample(x_train, y_train)
```

```python
modelLR = LogisticRegression()
modelLR.fit(X_train, Y_train)
y_predLR = modelLR.predict(x_test)
print(accuracy_score(y_test, y_predLR))
print(confusion_matrix(y_test, y_predLR))
print(f1_score(y_test, y_predLR, average='weighted'))
```

```python
modelRF = RandomForestClassifier()
modelRF.fit(X_train, Y_train)
y_predRF = modelRF.predict(x_test)
print(accuracy_score(y_test, y_predRF))
print(f1_score(y_test, y_predRF, average='weighted'))
```

```python
modelXG = XGBClassifier()
modelXG.fit(X_train, Y_train)
y_predXG = modelXG.predict(x_test)
print(accuracy_score(y_test, y_predXG))
print(f1_score(y_test, y_predXG, average='weighted'))
```

```python
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train_sc = scaler.fit_transform(x_train)
x_test_sc = scaler.transform(x_test)
modelSVC = SVC()
modelSVC.fit(x_train_sc, y_train)
y_predSVC = modelSVC.predict(x_test_sc)
```
