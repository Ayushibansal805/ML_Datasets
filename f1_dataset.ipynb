```python
import pandas as pd
df = pd.read_csv("/content/f1_strategy_dataset_v4.csv")
df
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['Driver'] = le.fit_transform(df['Driver'])
```

```python
df['Compound'] = le.fit_transform(df['Compound'])
```

```python
df['Race'] = le.fit_transform(df['Race'])
```

```python
df.info()
```

```python
y = df['Compound']
x = df.drop('Compound', axis=1)
```

```python
 from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=15)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
modelLR = LogisticRegression()
```

```python
modelLR.fit(x_train, y_train)
from sklearn.metrics import accuracy_score, confusion_matrix, f1_score
y_predLR = modelLR.predict(x_test)
print(accuracy_score(y_test, y_predLR))
print(confusion_matrix(y_test, y_predLR))
print(f1_score(y_test, y_predLR, average='weighted'))
```

```python
modelRF = RandomForestClassifier()
modelRF.fit(x_train, y_train)
y_predRF = modelRF.predict(x_test)
print(accuracy_score(y_test, y_predRF))
print(f1_score(y_test, y_predRF, average='weighted'))
```

```python
modelXG = XGBClassifier()
modelXG.fit(x_train, y_train)
y_predXG = modelXG.predict(x_test)
print(accuracy_score(y_test, y_predXG))
print(f1_score(y_test, y_predXG, average='weighted'))
```

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE()
X_train, Y_train = smote.fit_resample(x_train, y_train)
```

```python
modelLR = LogisticRegression()
modelLR.fit(X_train, Y_train)
y_predLR = modelLR.predict(x_test)
print(accuracy_score(y_test, y_predLR))
print(confusion_matrix(y_test, y_predLR))
print(f1_score(y_test, y_predLR, average='weighted'))
```

```python
modelRF = RandomForestClassifier()
modelRF.fit(X_train, Y_train)
y_predRF = modelRF.predict(x_test)
print(accuracy_score(y_test, y_predRF))
print(f1_score(y_test, y_predRF, average='weighted'))
```

```python
modelXG = XGBClassifier()
modelXG.fit(X_train, Y_train)
y_predXG = modelXG.predict(x_test)
print(accuracy_score(y_test, y_predXG))
print(f1_score(y_test, y_predXG, average='weighted'))
```

```python
from sklearn.svm import SVC #without smote
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train_sc = scaler.fit_transform(x_train)
x_test_sc = scaler.transform(x_test)
modelSVC = SVC()
modelSVC.fit(x_train_sc, y_train)
y_predSVC = modelSVC.predict(x_test_sc)
```

```python
acc_SVM = accuracy_score(y_test, y_predSVC)
print(acc_SVM)
print(f1_score(y_test, y_predSVC, average='weighted'))
```
